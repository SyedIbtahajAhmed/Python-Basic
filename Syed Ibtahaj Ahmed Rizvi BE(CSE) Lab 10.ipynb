#### Question No.1
You work in the IT department of an amusement park. You have been assigned a task to create
a program helps the security guards to allow only those people to enter who satisfy the
following conditions:
        - The visitors' age must be between 6 and 50 years old
        - The visitor height must be at least 2.5 feet
        - The visitor should must not be carrying any weapons
If the visitor fails to fulfill all or any of the three conditions, the program should display
"ENTRY NOT ALLOWED" so the guards can keep the visitors out. If the visitor fulfills the
requirements, the program should display "YOU MAY ENTER".


```python
age = float(input("Enter Your Age :"))
height = float(input("Enter Your Height in feet :"))
weapon = str(input("Is The Visitor Carrying Weapons :"))
if(age>=6 and age<=50 and height>=2.5 and weapon=="no" or weapon=="No"):
    print("YOU MAY ENTER")
else:
    print("ENTRY NOT ALLOWED")
```

#### Question No. 2
You are required to prepare the result of Computing Fundamentals course. Write a program that
fulfils the following conditions:
  - The passing marks are 60 out of 100 for lab course
  - The passing marks are 50 out of 100 for theory
  - If the student passes in both, the program should display "YOU PASS"
  - If the student fails in any one or both, the program should display "YOU FAILED"


```python
lab = float(input("Enter Marks Of Computing Fundamentals Lab :"))
theory = float(input("Enter Marks Of Computing Fundamentals Theory :"))
if(lab>=60 and lab<=100 and theory>=50 and theory<=100):
    print("YOU PASSED")
else:
    print("YOU FAILED")
```

#### Question 4:

Implement MAJU's grading policy through a Python Code. The user should enter their marks (out of 100). The program should print the user's grade.

##### For Example:
If the user enters 83, the program should print "Your Grade is B+ with 3.33 Grade Points"

MAJU's grading policy is as follows:


```python
marks = float(input("Enter your marks out of 100:"))
if(marks>=90 and marks<=100):
    print("Your Grade is A with 4.00 Grade Points")
elif(marks>=86 and marks<90):
    print("Your Grade is A- with 3.67 Grade Points")
elif(marks>=81 and marks<86):
    print("Your Grade is B+ with 3.33 Grade Points")
elif(marks>=77 and marks<81):
    print("Your Grade is B with 3.00 Grade Points")
elif(marks>=72 and marks<77):
    print("Your Grade is B- with 2.67 Grade Points")
elif(marks>=68 and marks<72):
    print("Your Grade is C+ with 2.33 Grade Points")
elif(marks>=63 and marks<68):
    print("Your Grade is C with 2.00 Grade Points")
elif(marks>=58 and marks<63):
    print("Your Grade is C- with 1.67 Grade Points")
elif(marks>=54 and marks<58):
    print("Your Grade is D+ with 1.33 Grade Points")
elif(marks>=50 and marks<54):
    print("Your Grade is D with 1.00 Grade Points")
else:
    print("Your Grade is F with 0.00 Grade Points. You Failed!")
```
